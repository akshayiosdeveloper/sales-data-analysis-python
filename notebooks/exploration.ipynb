```python
import pandas as pd
```

```python
# Step 1 — Load Data
df = pd.read_csv('sales_data.csv')
df
```

```python
# step 2 Data cleaning 
df.dropna(inplace=True)
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# Step 3 — Analysis (Pandas + NumPy)
import numpy as np
total_sales = df['sales'].sum()
print(total_sales)
average = np.mean(df['profit'])
print(average)
```

```python
# Step 4 — Grouping (Real-world skill)
regional_sales = df.groupby('region')['sales'].sum()
print(regional_sales)
```

```python
# Step 5 — Visualization (🔥 Important)
# bar chart 
import matplotlib.pyplot as plt
#plt.plot(regional_sales)
regional_sales.plot(kind="bar")
plt.title('mini project')
plt.show()
```

```python
# Line Chart (time-based)
df.groupby('date')['sales'].sum().plot()
plt.title('sales over time')
plt.show()
```

```python
#Top Products
top_product = df.groupby('product')['sales'].sum().sort_values(ascending=False)
print(top_product)
```

```python
monthly_sales = df.groupby(df['date'].dt.month)['sales'].sum()
print(monthly_sales)
```
